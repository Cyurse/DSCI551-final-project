```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot
youtube = pd.read_csv("data_agg.csv")
youtube.head()
```

```python
youtube.info()
```

```python
youtube[['views','likes','dislikes','comment_count','followers','videos']].describe()
```

```python
youtube['category_name'].value_counts().head(10).sort_values().plot(kind='barh', figsize=(7,4))
matplotlib.pyplot.title("Top 10 Category of Videos", size=20)
matplotlib.pyplot.show()
```

```python
youtube['channel_title'].value_counts().head(10).sort_values().plot(kind='barh', figsize=(7,4))
matplotlib.pyplot.title("Top 10 Youtube Channels", size=20)
matplotlib.pyplot.show()
```

```python
youtube['likes_log'] = np.log(youtube['likes'] + 1)
fig, ax = matplotlib.pyplot.subplots(figsize=(5, 6))
ax = sns.boxplot(y="category_name", x='likes_log', data=youtube, orient="h")
matplotlib.pyplot.title("View counts by Category", size=20)
matplotlib.pyplot.show()
```

```python
corr_list = ['views','likes','dislikes','comment_count','followers','videos']
youtube[corr_list].corr()
```

```python
new_corr_list = ['views','likes','dislikes','comment_count']
corr = youtube[new_corr_list].corr()
matplotlib.pyplot.figure(figsize=(8,6))
sns.heatmap(corr, annot=True);
```
